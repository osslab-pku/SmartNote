```python
LABEL_COL = 'type'
# count distinct
NUM_LABELS = len(df_commits[LABEL_COL].unique())
# select columns starting with 'lang'
CATE_COLS = ['domain', 'releaseType']
# columns with type 'number' or 'bool'
NUM_COLS = df_commits.select_dtypes(include=['number', 'bool']).columns.tolist()
# drop isInRN from NUM_COLS
NUM_COLS.remove('isInRN')
```

```python
from transformers import AutoConfig, AutoModel, AutoTokenizer
from multimodal_transformers.model import AutoModelWithTabular
from multimodal_transformers.model import TabularConfig

settings.MODEL_NAME = 'baai/bge-large-en-v1.5'

model_config = AutoConfig.from_pretrained(settings.MODEL_NAME)
tokenizer = AutoTokenizer.from_pretrained(settings.MODEL_NAME)
model_config.tabular_config = TabularConfig(
        combine_feat_method='attention_on_cat_and_numerical_feats',  # change this to specify the method of combining tabular data
        cat_feat_dim=len(CATE_COLS),  # need to specify this
        numerical_feat_dim=len(NUM_COLS),  # need to specify this
        num_labels=NUM_LABELS,   # need to specify this, assuming our task is binary classification
        use_num_bn=False,
)

model = AutoModelWithTabular.from_pretrained(settings.MODEL_NAME, config=model_config).to(settings.TORCH_DEVICE)
model
```
