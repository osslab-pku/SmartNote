```python
def calculate_correlation(s1, s2):
    s1 = pd.Series(list(s1))
    s2 = pd.Series(list(s2))
    # drop if any of the values is NaN
    s1_nas = s1.isna()
    s2_nas = s2.isna()
    nas = s1_nas | s2_nas
    s1 = s1[~nas]
    s2 = s2[~nas]
    return s1.corr(s2, 'spearman')

df_auto = pd.read_csv('qual_eval.csv')
df_auto = df_auto.merge(pd.read_csv('qual_eval2.csv'))
df_auto_grouped = df_auto.groupby('rn_name')
df_auto_grouped
```

```python
_l = []
# preprocess human 
# calculate avg for each project
df_survey_grouped = df_survey.groupby('project')
# unfold the 0,1,2,3 -> rows of 0_original_rn, 1_smartdraft_rn_automatic, 2_deeprelease_rn, 3_conventional_changelog
for project, group in df_survey_grouped:
    _l.append({
        'project': project,
        'rn_name': '0_original_rn',
        'completeness': group['completeness_0'].mean(),
        'clarity': group['clarity_0'].mean(),
        'conciseness': group['conciseness_0'].mean(),
        'organization': group['organization_0'].mean()
    })
    _l.append({
        'project': project,
        'rn_name': '1_smartdraft_rn_automatic',
        'completeness': group['completeness_1'].mean(),
        'clarity': group['clarity_1'].mean(),
        'conciseness': group['conciseness_1'].mean(),
        'organization': group['organization_1'].mean()
    })
    _l.append({
        'project': project,
        'rn_name': '2_deeprelease_rn',
        'completeness': group['completeness_2'].mean(),
        'clarity': group['clarity_2'].mean(),
        'conciseness': group['conciseness_2'].mean(),
        'organization': group['organization_2'].mean()
    })
    _l.append({
        'project': project,
        'rn_name': '3_conventional_changelog',
        'completeness': group['completeness_3'].mean(),
        'clarity': group['clarity_3'].mean(),
        'conciseness': group['conciseness_3'].mean(),
        'organization': group['organization_3'].mean()
    })
    
df_survey_avg = pd.DataFrame(_l)
df_survey_avg 
```

```python
# transform df_auto
df_auto['project'] = df_auto['repo_name'].apply(lambda x: x.split('/')[-1])
df_auto.drop(columns=['repo_name'], inplace=True)
df_auto = df_auto[df_auto['rn_name'].isin(['0_original_rn', '1_smartdraft_rn_automatic', '2_deeprelease_rn', '3_conventional_changelog'])]
df_auto = df_auto.merge(df_survey_avg, on=['project', 'rn_name'])
df_auto.dropna(inplace=True)
df_auto
```
